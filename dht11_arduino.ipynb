# **Projetinho Arduino**
## **Sensor DHT11**

O projetinho teve como objetivo aprender um pouco mais de análise de dados e se divertir um pouco, utilizando Arduino com o sensor DHT11 para a extração de dados (Data e hora, temperatura e umidade) e o Python como a linguagem de programação para fazer o estudo destes dados obtidos durante 4,5 dias.

```python
#Importando as bibliotecas necessárias

import pandas, seaborn
import matplotlib.pyplot as plt
```

**Análise Geral (de todos os dias)**

```python
#Carregando a planilha aqui, por enquanto apenas a análise geral

arduino_data = pandas.read_excel('arduino_teste.xlsx', sheet_name = 'Original')
arduino_data.head()
```

```python
#Descrição dos dados

arduino_data.describe().T
```

Na descrição estatística que o python nos fornece, temos a contagem das linhas, média, desvio padrão, valores mínimos e máximos.

Olhando também para os valores que mais ocuparam a planilha, conclui-se que os 4,5 dias que se passaram, eram dias quentes mas também com uma boa chance de chuva.

```python
#Realizando os gráficos para melhor análise
#Data e hora x Umidade
plt.figure(figsize = (10, 6))
plt.plot(arduino_data['TIME'], arduino_data['Umidade (%)'])
plt.title('Data e Hora x Umidade')
plt.xlabel('Data e Hora')
plt.ylabel('Umidade (%)')
```

A partir do gráfico acima podemos concluir que a umidade era um pouco mais baixa durante o pico do dia (meio dia e um pouco da tarde) e alta no começo da noite até o fim da manhã.

```python
#Data e Hora x Temperatura
plt.figure(figsize = (10, 6))
plt.scatter(arduino_data['TIME'], arduino_data['Temperatura (ºC)'])
plt.title('Data e Hora x Temperatura')
plt.xlabel('Data e Hora')
plt.ylabel('Temperatura (ºC)')
```

Neste gráfico, podemos perceber que em grande parte dos dias a temperatura se manteu entre 26 e 27 graus.
